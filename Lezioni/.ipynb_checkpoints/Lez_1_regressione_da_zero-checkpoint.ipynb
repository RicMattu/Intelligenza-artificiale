La costruiremo in modi diversi procedendo per difficoltà crescenti:

### Livello 1.1 
- Dataset con dimensioni ridotte  
- Regressione 1D

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
# Dataset
x = np.linspacee(0,10,50)
y_t = np.pi * x**2 + np.e     # ground truth
```

```python
# Splitting in training and test set 80-20
x_train = x[0:40]
x_test = x[40:]

y_train = y_t[0:40]
y_test = y_t[40:]
```

```python
# Definisco la funzione di attivazione
def s(x):
    return 1/(1+np.e**(-x))

def d_s(x):
    return s(x)*(1-s(x))
```

Definisco la rete come un semplice MLP con 1 hidden layer:  
$$
\hat{y} = MLP(x) = \sigma(x \cdot w_{0} + b_{0}) \cdot w_1 + b_1
$$

```python
# Definisco la rete: MLP con 1 h.l.

## Inizializzo il vettore dei pesi e dei bias
w = 10*np.random.rand(2) 
b = w

def MLP(x):
    h = s(w[0]*x + b[0])  
    y_pred = w[1]*h + b[1]
    return y_pred
```

```python
# Definisco la loss (MSE)   
def loss(y_t,y_pred):
    err = np.mean((y_t-y_pred)**2)
    return err
```

```python
# Stampo la loss finale
print(f"Final loss: {L[-1]}")
```

```python
# Stampo il grafico della loss nel training
#%matplotlib inline
plt.figure(figsize=(6,3))
plt.plot(range(epochs), L)
plt.show()    
```

```python
# Stampo la loss nel test set:

# Forward pass
y_pred_test = MLP(x_test)
L.append(loss(y_t_test,y_pred_test))
print(f"Test Loss: {L[0]:.4f}")
```
