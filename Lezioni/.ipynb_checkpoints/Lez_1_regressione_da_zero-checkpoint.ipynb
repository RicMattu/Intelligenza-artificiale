### Livello 1.1

````python
# Splitting in training and test set 80-20
x_train = x[0:40]
x_test = x[40:]

y_train = y_t[0:40]
y_test = y_t[40:]
````

````python
# Definisco la funzione di attivazione
def s(x):
    return 1/(1+np.e**(-x))

def d_s(x):
    return s(x)*(1-s(x))
````

Definisco la rete come un semplice MLP con 1 hidden layer:  
$$
\hat{y} = MLP(x) = \sigma(x \cdot w_{0} + b_{0}) \cdot w_1 + b_1
$$

````python
# Definisco la rete: MLP con 1 h.l.

## Inizializzo il vettore dei pesi e dei bias
w = 10*np.random.rand(2) 
b = w

def MLP(x):
    h = s(w[0]*x + b[0])  
    y_pred = w[1]*h + b[1]
    return y_pred
````

````python
# Definisco la loss (MSE)   
def loss(y_t,y_pred):
    err = np.mean((y_t-y_pred)**2)
    return err
````

### Livello 1.2
- Dataset con dimensioni ridotte  
- Regressione ND
- MLP con numero di h_l a piacere

Cioè supponiamo che la relazione input-output sia della forma
$$ y = f(x_1,\dots,x_n) $$

In questo caso definiamo la MLP come una classe di python (ereditata da torch.nn.Module),  
questa scelta ci permette di specificare da subito quali sono i paramtetri trainable del modello e di gestire più facilmente il backward.

````python
# importo le solite librerie più torch.nn
import numpy as np
import matplotlib.pyplot as plt
import torch #uso pytorch solo per gestire le matrici
import torch.nn as nn
np.random.seed(0)
````

````python
# Iperparametri
n = 4  # input dimension
if (n%2 != 0):
    print("Ho bisogno di n pari quindi sto usando n+1\n")
````

````python
# Dataset
x = torch.rand((50,n))
y = np.pi * x[0][0]**2 * np.exp(x[0][1]) + torch.sum(x,1)

x_train = x[:40]
y_train = y[:40]

x_test = x[40:]
y_test = y[40:]
````

Costruendo così la rete in pratica cosa succede:
$$
x \in \mathbb{R}^n \mapsto H_1 \cdot x =: x_{(1)} \in \mathbb{R}^{(n/2)}
$$
E ad ogni ierazione
$$
x_{(i)} \in \mathbb{R}^{(n/2)} \mapsto H_i \cdot x_{(i)} =: x_{(i+1)} \in \mathbb{R}^{(n/i+1)}
$$

````python
class MLP(nn.Module):
    def __init__(self,n,num_hid = 3):
        super().__init__()

        # Parametri
        self.num_hid = num_hid
        
        # Learneble parameters
        self.weight_matrix = nn.ParameterList()
        self.bias          = nn.ParameterList()
        
        # Build the Net
        # 1st layer: n -> n/2
        in_dim = n
        out_dim = round(n/2)
        self.weight_matrix.append(nn.Parameter(torch.rand(in_dim,out_dim))) #(n,n/2)
        self.bias.append(nn.Parameter(torch.rand(out_dim)))                  # (n/2)
        
        # hidden layers   
        
        for i in range(self.num_hid):
            in_dim = out_dim
            out_dim = round(n/(i+2))
            self.weight_matrix.append(nn.Parameter(torch.rand(in_dim, out_dim))) #(n/2,n/3)...
            self.bias.append(nn.Parameter(torch.rand(out_dim)))
            
        # ouptut layer  
        self.weight_matrix.append(nn.Parameter(torch.rand(out_dim,1)))
        self.bias.append(nn.Parameter(torch.rand(1)))
                
    def forward(self,x):        
        h = x
        for W,b in zip(self.weight_matrix[:-1],self.bias[:-1]):
            h =  torch.sigmoid(h @ W + b)
            
        y = h @ self.weight_matrix[-1] + self.bias[-1]     
        
        return y


# inizializzo
net = MLP(n)

# definisco la loss
loss_fn = nn.MSELoss()
````

````python
# Forward pass 
y_pred = net(x_train)
y_train = y_train.squeeze()
loss = loss_fn(y_train,y_pred)
loss
````

````python







# Forward pass
y_pred = net(x_train)
loss = loss_fn(y_train,y_pred)
print(f"Initial Loss: {loss:.4f}")

# Training
epochs = 25
L = []
optimizer = torch.optim.SGD(net.parameters(),lr=0.01)

for i in range(epochs):
    # forward pass
    y_pred = net(x_train)
    loss = loss_fn(y_train,y_pred)
    
    # resetto il gradiente
    optimizer.zero_grad()
    
    # calcolo i gradienti
    loss.backward()
    L.append(loss.item())
    
    # update weights
    optimizer.step()
    
# Stampo la loss finale
#print(f"Final loss: {L[-1]}")

# Stampo il grafico della loss nel training
#%matplotlib inline
plt.figure(figsize=(6,3))
plt.plot(range(epochs), L)
plt.show() 



````

````python
class MLP2(nn.Module):
    def __init__(self,n,num_hid = 3):
        super().__init__()
        
        layers = []
        
        # self.weight_matrix = nn.ParameterList()
        # self.bias = nn.ParameterList()
        # self.num_hid = num_hid
        
        # 1st layer: n -> n/2
        layers.append(nn.Linear(n,round(n/2)))
        layers.append(nn.Sigmoid())

        
        # hidden layers   
        out_dim = round(n/2)
        for i in range(num_hid-1):
            layers.append(nn.Linear(out_dim,round(n/(i+3))))
            layers.append(nn.Sigmoid())
            out_dim = round(n/(i+3))

        # ouptut layer  
        layers.append(nn.Linear(out_dim,1))
        
        self.for_pass = nn.Sequential(*layers)
                
    def forward(self,x):  
        return self.for_pass(x)
````
