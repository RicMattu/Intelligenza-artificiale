L'obiettivo di questo notebook è mostrare come si può costruire una rete neurale (MLP) che faccia una regressione.  
Le attuali librerie in python consentono di ridurre le linee di codice per un task simile a meno di 50 righe, sfruttando classi e funzioni già definite.

Tuttavia, almeno all'inizio, anche per capire a pieno cosa si sta usando quando si sfruttano determinate librerie, è - a mio avviso- opportuno costruirla dal livello più basso possibile. Perciò costruiremo un MLP in modi diversi procedendo per difficoltà crescenti:
1. la costruiamo da zero usando solo operazioni algebriche
2. usiamo Sequential e automatizziamo i gradienti
3. usiamo la libreria torch

### Livello 1
In questo caso importiamo solo due librerie, numpy per le operazioni matematiche e matplotlib per i plot. 
Inoltre, senza perdita di generalità, useremo un dataset con dimensioni ridotte e ci concentreremo, per semplicità, su una regressione 1D

````python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0) # per la riproducibilità
````

````python
# Dataset
x = np.linspace(0,10,100)
y_t = np.pi * x**2 + np.e     # ground truth
````

````python
# Splitting in training and test set 80-20
x_train = x[0:60]
x_test = x[60:]

y_train = y_t[0:60]
y_test = y_t[60:]
````

````python
# Definisco la funzione di attivazione
def s(x):
    return 1/(1+np.e**(-x))

def d_s(x):
    return s(x)*(1-s(x))
````

Definisco la rete come un semplice MLP con 1 hidden layer:  
$$
\hat{y} = MLP(x) = \sigma(x \cdot w_{0} + b_{0}) \cdot w_1 + b_1
$$

````python
# Definisco la rete: MLP con 1 h.l.

## Inizializzo il vettore dei pesi e dei bias
w = 10*np.random.rand(2) 
b = w

def MLP(x):
    h = s(w[0]*x + b[0])  
    y_pred = w[1]*h + b[1]
    return y_pred
````

````python
# Definisco la loss (MSE)   
def loss(y_t,y_pred):
    err = np.mean((y_t-y_pred)**2)
    return err
````

````python
# Inizializzo la loss per memorizzarla
L = []
````

````python
# Forward pass
y_pred = MLP(x_train)
L.append(loss(y_train,y_pred))
print(f"Initial Loss: {L[0]:.4f}")
````

````python
# Training (Backpropagation)
L = []
epochs = 10
lr = 0.1

for epoch in range(epochs):
    # forward
    y_pred = MLP(x_train)
    
    # compute loss
    L.append(loss(y_train,y_pred))
    print(f"Loss at epoch {epoch+1} = {L[epoch]:.3f}")
    
    # compute gradients
    grad_w = np.zeros(2)
    grad_b = np.zeros(2)
    
    h = s(w[0]*x_train + b[0]) 
    
    grad_w[0] = 2*np.mean((y_pred-y_train)*w[1]*d_s(h)*x_train)
    grad_w[1] = 2*np.mean((y_pred-y_train)*h)   
    grad_b[0] = 2*np.mean((y_pred-y_train)*w[1] * d_s(w[0]*x_train + b[0]))
    grad_b[1] = 2*np.mean(y_pred-y_train)    
    grad_w = np.array(grad_w)
    grad_b = np.array(grad_b)
    
    # update wights
    w = w - lr*grad_w
    b = b - lr*grad_b
````

````python
# Stampo la loss finale
print(f"Final loss: {L[-1]:.3}\n")
print(f"I pesi finali sono w_0 = {w[0]:.2f}, b_0 = {b[0]:.2f}, w_1 = {w[1]:.2f}, b_1 = {b[1]:.2f}\n")
````

````python
# Visualizzazione risultati
def pred_mod(x):
    return s(w[0]*x+b[0])*w[1] + b[1]

pred_res = MLP(x_test)    
plt.figure(figsize=(6,3))
plt.plot(x_test,y_test, color = 'green', label = 'ground truth')
plt.plot(x_test,pred_res, marker = '.', color = 'red', label = 'prediction'),
plt.legend()
plt.show() 
````

E' normale che il risultato faccia un po' schifo, questo è dovuto alla semplicità della rete e alla scelta della sigmoide come funzione di attivazione, in ogni caso si può vedere che il training ha un effetto positivo

````python
# Stampo il grafico della loss nel training
#%matplotlib inline
plt.figure(figsize=(6,3))
plt.plot(range(epochs), L)
plt.show()    
````

````python
# Stampo la loss nel test set:

# Forward pass
y_pred_test = MLP(x_test)
L.append(loss(y_test,y_pred_test))
print(f"Test Loss: {L[0]:.4f}")
````

### Livello 1.2
Ora utilizziamo la libreria torch, ma solo per gestire meglio le matrici e consideriamo una regressione N-dimensionale:
cioè supponiamo che la relazione input-output sia della forma
$$ y = f(x_1,\dots,x_n) $$

In questo caso definiamo la MLP come una classe di python (ereditata da torch.nn.Module),  
questa scelta ci permette di specificare da subito quali sono i paramtetri trainable del modello e di gestire più facilmente il backward, inoltre di specificare a piacere il numero di hidden layer.

````python
# importo le solite librerie più torch.nn
import numpy as np
import matplotlib.pyplot as plt
import torch #uso pytorch solo per gestire le matrici
import torch.nn as nn
np.random.seed(0)
````

````python
# Iperparametri
n = 2  # input dimension
if (n%2 != 0):
    print("Ho bisogno di n pari quindi sto usando n+1\n")
````

````python
# Dataset
x = torch.rand((100,n))
y = np.pi * x[0][0]**2 * np.exp(x[0][1]) + torch.sum(x,1)

x_train = x[:60]
y_train = y[:60]

x_test = x[60:]
y_test = y[60:]
````

Costruendo così la rete in pratica cosa succede:
$$
x \in \mathbb{R}^n \mapsto H_1 \cdot x + b_1  =: x_{(1)} \in \mathbb{R}^{(n/2)}
$$
E ad ogni ierazione
$$
x_{(i)} \in \mathbb{R}^{(n/2)} \mapsto H_i \cdot x_{(i)} + b_i =: x_{(i+1)} \in \mathbb{R}^{(n/i+1)}
$$

````python
class MLP(nn.Module):
    def __init__(self,n,num_hid = 3):
        super().__init__()

        # Parametri
        self.num_hid = num_hid
        
        # Learneble parameters
        self.weight_matrix = nn.ParameterList()
        self.bias          = nn.ParameterList()
        
        # Build the Net
        # 1st layer: n -> n/2
        in_dim = n
        out_dim = round(n/2)
        self.weight_matrix.append(nn.Parameter(torch.rand(in_dim,out_dim))) #(n,n/2)
        self.bias.append(nn.Parameter(torch.rand(out_dim)))                  # (n/2)
        
        # hidden layers   
        
        for i in range(self.num_hid):
            in_dim = out_dim
            out_dim = round(n/(i+2))
            self.weight_matrix.append(nn.Parameter(torch.rand(in_dim, out_dim))) #(n/2,n/3)...
            self.bias.append(nn.Parameter(torch.rand(out_dim)))
            
        # ouptut layer  
        self.weight_matrix.append(nn.Parameter(torch.rand(out_dim,1)))
        self.bias.append(nn.Parameter(torch.rand(1)))
                
    def forward(self,x):        
        h = x
        for W,b in zip(self.weight_matrix[:-1],self.bias[:-1]):
            h =  torch.sigmoid(h @ W + b)
            
        y = h @ self.weight_matrix[-1] + self.bias[-1]     
        
        return y


# inizializzo
net = MLP(n)

# definisco la loss
loss_fn = nn.MSELoss()
````

````python
# Forward pass 
y_pred = net(x_train)
y_pred = y_pred.squeeze()
loss = loss_fn(y_train,y_pred)
print(f"Initial Loss: {loss:.4f}")
````

````python
# Training
epochs = 25
L = []
optimizer = torch.optim.SGD(net.parameters(),lr=0.01)

for i in range(epochs):
    # forward pass
    y_pred = net(x_train).squeeze()
    loss = loss_fn(y_train,y_pred)
    
    # resetto il gradiente
    optimizer.zero_grad()
    
    # calcolo i gradienti e la loss 
    loss.backward()
    L.append(loss.item())
    
    # update weights
    optimizer.step()
````

````python
# Stampo la loss finale
print(f"Final loss: {L[-1]}")

# Stampo il grafico della loss nel training
#%matplotlib inline
plt.figure(figsize=(6,3))
plt.plot(range(epochs), L)
plt.show() 
````

````python
class MLP2(nn.Module):
    def __init__(self,n,num_hid = 3):
        super().__init__()
        
        layers = []
        
        # self.weight_matrix = nn.ParameterList()
        # self.bias = nn.ParameterList()
        # self.num_hid = num_hid
        
        # 1st layer: n -> n/2
        layers.append(nn.Linear(n,round(n/2)))
        layers.append(nn.Sigmoid())

        
        # hidden layers   
        out_dim = round(n/2)
        for i in range(num_hid-1):
            layers.append(nn.Linear(out_dim,round(n/(i+3))))
            layers.append(nn.Sigmoid())
            out_dim = round(n/(i+3))

        # ouptut layer  
        layers.append(nn.Linear(out_dim,1))
        
        self.for_pass = nn.Sequential(*layers)
                
    def forward(self,x):  
        return self.for_pass(x)
````

````python
import torch
import torch.nn as nn

class MLP(nn.Module):
    def __init__(self,in_dim):
        super().__init__()
        self.in_dim = in_dim
        out_dim = 10
        
        self.rete = nn.Sequential(
            nn.Linear(self.in_dim, out_dim),
            nn.ReLU(),
            nn.Linear(out_dim,1),
        )

    def forward(self,x):
        return self.rete(x)
````

````python
# Simple script for a test

x_train = torch.rand((100,1))*10
x_test = torch.rand((30,1))*10

def f(x):
    y1 = torch.sin(x)*5 + torch.pi*x - torch.sqrt(x)
    return y1 #torch.stack([y1, y2], dim=1)

y_train = f(x_train)
y_test = f(x_test)

mlp = MLP(1)
````

````python
# training
loss = nn.MSELoss()
optimizer = torch.optim.Adam(mlp.parameters(), lr = 0.01)
epochs = 250
L = []
y_train.shape
for i in range(epochs):
    
    # forward pass
    y_hat = mlp(x_train)
    
    err = loss(y_hat,y_train) # calcolo l'errore

    optimizer.zero_grad()     # resetto i grad.
    err.backward()            # backprop.
    optimizer.step()          # aggiorno
    L.append(err.item())
    print(f"Epoch: {i}, Loss: {err.item():.3}")
    
````

````python
pred_res = mlp(x_test).detach().numpy()


plt.figure(figsize=(6,3))
plt.scatter(x_test, y_test, color='green', label='ground truth')
plt.scatter(x_test, pred_res, color='red', label='prediction')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
````
